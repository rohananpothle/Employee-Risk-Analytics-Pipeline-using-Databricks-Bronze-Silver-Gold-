## Gold Layer - Reporting Summary
**Purpose**: Create final reporting/summary table (Gold layer)

```python
from pyspark.sql.functions import count

# Read Silver table
clean_df = spark.table("rohan_catalog.benefits_demo.employee_cleaned")

# Aggregate risk summary
final_df = clean_df.groupBy("RiskCategory").agg(count("*").alias("TotalEmployees"))

# Save Gold table
final_df.write.mode("overwrite").saveAsTable("rohan_catalog.benefits_demo.risk_summary")

print("Gold layer summary refreshed successfully!")
```
