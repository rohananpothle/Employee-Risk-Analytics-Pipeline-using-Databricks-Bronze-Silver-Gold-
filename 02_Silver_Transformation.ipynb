## Silver Layer - Cleaned Employee Data
**Purpose**:Transform Bronze raw data into Silver cleaned business-ready table

```python
from pyspark.sql.functions import when

# Read Bronze table
df = spark.table("rohan_catalog.benefits_demo.employee_data")

# Keep only active employees
clean_df = df.filter("EnrollmentStatus = 'Active'")

# Add RiskCategory based on Smoking or Diabetic
clean_df = clean_df.withColumn(
    "RiskCategory",
    when((clean_df.Smoking == "Y") | (clean_df.Diabetic == "Y"), "High Risk")
    .otherwise("Low Risk")
)

# Drop metadata columns before writing to Silver
clean_df = clean_df.drop("LoadTimestamp", "SourceFile")

# Save Silver table
clean_df.write.mode("overwrite").saveAsTable("rohan_catalog.benefits_demo.employee_cleaned")

print("Silver layer refreshed successfully!")
```
